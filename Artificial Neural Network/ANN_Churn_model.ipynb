# Import necessary libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import tensorflow as tf
```

```python
# # Load dataset from the specified path

path = "/content/Churn_Modelling.csv"
df = pd.read_csv(path)
df
```

```python
# Split the independent features and label

X = df.iloc[:,3:13]
y = df.iloc[:,13]
```

```python
# Checking if we have any categorical data

categorical_data = [var for var in X.columns if X[var].dtypes == "O"]
categorical_data
```

```python
X["Geography"].value_counts()
```

```python
X["Gender"].value_counts()
```

```python
# So, let's convert this categorical data in Numerical data

geography = pd.get_dummies(X["Geography"], drop_first = True)
gender = pd.get_dummies(X["Gender"], drop_first = True)
```

```python
X = pd.concat([X, geography, gender], axis = 1)
```

```python
X.head()
```

```python
# dropping the Categorical cols

X.drop(categorical_data, axis = 1, inplace = True)
```

```python
X.head()
```

```python
# Splitting the dataset into train set and test set

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state=123)
```

```python
# feature Scaling

from sklearn.preprocessing import StandardScaler
sc= StandardScaler()
```

```python
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
X_train_pt = X_train.copy()
X_test_pt = X_test.copy()
```

```python
X_train.shape, X_test.shape
```

Working with ANN

```python
# Import required modules from Keras

import keras
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import LeakyReLU, ReLU, ELU, Dropout
```

```python
# Initailizing sequential model
clf = Sequential()
```

```python
# adding layers to the model

clf.add(Dense(units = 32, activation = "relu", kernel_initializer="he_uniform", input_dim=11))
clf.add(Dense(units = 16, activation = "relu", kernel_initializer="he_uniform"))
clf.add(Dense(units = 1, activation = "sigmoid", kernel_initializer="glorot_uniform"))
```

```python
# model compilation

clf.compile(optimizer = "Adam", loss="binary_crossentropy", metrics = ["accuracy"])
```

```python
# train the model on training dataset

model_history = clf.fit(X_train, y_train, validation_split=0.33, batch_size =20 , verbose=2, epochs=100)
```

```python
y_pred = clf.predict(X_test)
y_pred = (y_pred>0.5)
```

```python
# model evaluation

from sklearn.metrics import confusion_matrix, accuracy_score
cm = confusion_matrix(y_test, y_pred)
```

```python
print(cm)
```

```python
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy Score:", accuracy)
```

Now Hyperparameter tunning on ANN

```python
## Perform Hyperparameter Optimization

from keras.wrappers.scikit_learn import KerasClassifier
from sklearn.model_selection import GridSearchCV
from keras.models import Sequential
from keras.layers import Dropout
from keras.layers import Dense, Activation, ReLU, LeakyReLU
from keras.activations import relu, sigmoid
```

```python
# Define a function to create a Keras model with customizable architecture

def create_model(layers, activation):
    model = Sequential()
    for i, nodes in enumerate(layers):
        if i==0:
            model.add(Dense(nodes,input_dim=X_train.shape[1]))
            model.add(Activation(activation))
            model.add(Dropout(0.3))
        else:
            model.add(Dense(nodes))
            model.add(Activation(activation))
            model.add(Dropout(0.3))

    model.add(Dense(units = 1, kernel_initializer= 'glorot_uniform', activation = 'sigmoid')) # Note: no activation beyond this point

    model.compile(optimizer='adam', loss='binary_crossentropy',metrics=['accuracy'])
    return model
```

```python
# Creating a kerasClassifier from scikeras

# model = KerasClassifier(build_fn=create_model, verbose=2)

from scikeras.wrappers import KerasClassifier


model = KerasClassifier(
    model=create_model,
    verbose=2,
    activation= "relu",
    layers = 20
)
```

```python
# Perform grid search cv

layers = [[40, 20], [40,20,5],[64, 32, 16]]
activations = ['sigmoid', 'relu']


param_grid = dict(layers=layers, activation=activations, batch_size = [20, 50], epochs=[50])

grid = GridSearchCV(estimator=model, param_grid=param_grid,cv=5)
```

```python
grid_result = grid.fit(X_train_pt, y_train)
```

```python
# Best score and Best parameters for this model using scikeras

[grid_result.best_score_,grid_result.best_params_]
```
