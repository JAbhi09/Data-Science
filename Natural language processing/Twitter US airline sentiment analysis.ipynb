```python
# Code preprocesses text data for NLP tasks, expanding abbreviations and cleaning.

abbreviations = {'ppl': 'people','cust':'customer','serv':'service','mins':'minutes','hrs':'hours','svc': 'service',
           'u':'you','pls':'please'}

train_index = train[~train.negativereason_gold.isna()].index
# test_index = test[~test.negativereason_gold.isna()].index

for index, row in train.iterrows():
    tweet = row.text
    tweet = re.sub('((www\.[^\s]+)|(https?://[^\s]+))','',tweet) #remove links
    tweet = re.sub('@[^\s]+','',tweet) #remove usernames
    tweet = re.sub('[\s]+', ' ', tweet) #remove additional whitespaces
    tweet = re.sub(r'#([^\s]+)', r'\1', tweet) #replace '#' word with word after "#" eg--->(#coding -----> coding)
    tweet = tweet.strip('\'"') #trim tweet
    words = []
    for word in tweet.split():
        if word.lower() not in stops:
            if word in list(abbreviations.keys()):
                words.append(abbreviations[word])
            else:
                words.append(word.lower())   
    tweet = " ".join(words)
    tweet = " %s %s" % (tweet, row.airline)
    row.text = tweet
    if index in train_index:
        row.text = " %s %s" % (row.text, row.negativereason_gold)
```

### Creating vocab and data formatting

```python
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.svm import SVC
```

```python
X = train.drop("airline_sentiment", axis = 1)
y = train["airline_sentiment"]
```

```python
X
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=123)
```

### Tf-Idf

```python
tf_idf = TfidfVectorizer(analyzer='word', max_features=3150, max_df = 0.7, ngram_range=(1,3))
train_features= tf_idf.fit_transform(X_train["clean_text"])
test_features=tf_idf.transform(X_test["clean_text"])
```

```python
for feature_name in tf_idf.get_feature_names_out():
    print(feature_name)
```

```python
# clf = SVC(kernel="linear", C= 0.96 , gamma = 'scale')
# clf = SVC(C = 1000, gamma = 0.001)
clf = SVC(kernel="linear", C= 0.9 , gamma = 'scale')
clf.fit(train_features, y_train)
# pred = clf.predict(test_features)
```

```python
clf.score(train_features, y_train)
```

```python
clf.score(test_features, y_test)
```
